```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
trump = '/data/parush/stance_mohammed/trump.txt'
```

```python
df = pd.read_csv(trump, sep="\t")
X = df[['ID','Target','Tweet']]
y = df[['Stance']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True,stratify = y)
```

```python
X_train['Stance'] = y_train
X_test['Stance'] = y_test
```

```python
train_data_file_m = '/data/parush/stance_mohammed/train.txt'
test_data_file_m = '/data/parush/stance_mohammed/test.txt'
```

```python
df_train = pd.read_csv(train_data_file_m, sep="\t")
```

```python
df_train = pd.concat([df_train, X_train], ignore_index=True)
df_train
```
